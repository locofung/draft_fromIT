```python
df_staff = pd.DataFrame({'First Name': ['Aryan', 'Rohan', 'Riya', 'Yash', 'Siddhant', ],
                   'Last Name': ['Singh', 'Agarwal', 'Shah', 'Bhatia', 'Khanna'],
                   'Type': ['Full-time Employee', 'Intern', 'Full-time Employee', 
                            'Part-time Employee', 'Full-time Employee'],
                   'Department': ['Administration', 'Technical', 'Administration', 
                                  'Technical', 'Management'],
                   'YoE': [2, 3, 5, 7, 6],
                   'Salary': [20000, 5000, 10000, 10000, 20000]})

df_staff
```

```python
staff_table=pd.pivot_table(data=df_staff,index=['Type'],columns=['Department'],values=['Salary'])
staff_table
```

```python
staff_agg=pd.pivot_table(data=df_staff,index=['Type'],columns=['Department'],values=['Salary']
                        ,aggfunc=['sum'])
staff_agg
```

```python
"""
find each depts salary info :sum,count,mean
"""
staff_aggs=pd.pivot_table(data=df_staff,index=['Type'],columns=['Department'],values=['Salary']
                        ,aggfunc=['sum','count','mean'])
staff_aggs
```

```python
"""
find whole company salary info(no need to precise to dept level)
"""
staff_aggs=pd.pivot_table(data=df_staff,index=['Type'],values=['Salary']
                        ,aggfunc=['sum','count','mean'])
staff_aggs
```

```python
"""
drill down to dept level,but more reader-friendly
"""
staff_aggs2=pd.pivot_table(data=df_staff,index=['Type','Department'],values=['Salary']
                        ,aggfunc=['sum','count','mean'])
staff_aggs2
```
